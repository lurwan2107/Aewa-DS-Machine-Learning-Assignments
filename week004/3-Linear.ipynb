```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn import datasets, linear_model, model_selection
```

```python
m, n = datasets.load_diabetes(return_X_y=True)
print(m.shape)
print(m[1])
```

```python
m = m[:, np.newaxis, 0]

m_train, m_test, n_train, n_test = model_selection.train_test_split(m, n, test_size=0.33)
```

```python
train = linear_model.LinearRegression()
train.fit(m_train, n_train)
```

```python
predict = train.predict(m_test)
```

```python
plt.scatter(m_test, n_test,  color='green')
plt.plot(m_test, predict, color='red', linewidth=3)
plt.show()
```
