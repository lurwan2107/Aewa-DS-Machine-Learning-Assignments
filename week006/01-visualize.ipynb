```python
import matplotlib.pyplot as plt 
import pandas as pd 
song = pd.read_csv('C:/Users/PC/Desktop/Arewa-DS-Machine-Learning-Assignments/nigerian-song.csv')
song.head()
```

```python
song.isnull().sum()
```

```python
top = song['artist_top_genre'].value_counts()
plt.figure(figsize=(6,6))
plt.scatter(x=top[:6].index, y=top[:6].values)
plt.xticks(rotation=45)
plt.title('Top genre', color='blue')
plt.show()
```

```python
missin = song[song['artist_top_genre'] != 'Missing']
toop = song['artist_top_genre'].value_counts()
plt.figure(figsize=(8,8))
plt.scatter(x=toop.index, y=toop.values)
plt.xticks(rotation=60)
plt.title('Top geners', color='red')
plt.show()
```
