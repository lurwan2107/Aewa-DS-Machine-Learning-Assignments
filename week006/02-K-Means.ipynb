Trying Mean-Shift Clustering Method:

```python
# Importing the required libraries
from numpy import unique
from numpy import where
import matplotlib.pyplot as plt 
from sklearn.datasets import make_classification
from sklearn.cluster import DBSCAN
from sklearn.mixture import GaussianMixture
from sklearn.cluster import MeanShift
```

```python
# initialize the data set we'll work with
training_data, _ = make_classification(
    n_samples=1000,
    n_features=2,
    n_informative=2,
    n_redundant=0,
    n_clusters_per_class=1,
    random_state=4
)
```

```python
# define the model for training
model = MeanShift()
result = model.fit_predict(training_data)
clusters = unique(result)
```

```python
# plotting the  Mean-Shift clustering
for cluster in clusters:
    index = where(result == cluster)
    plt.scatter(training_data[index, 0], training_data[index, 1])

plt.show()
```
