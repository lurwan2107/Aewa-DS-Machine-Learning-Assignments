In this assignment i created my own dataset to explore more interesting visualization of simple line chart and bar chart

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
population = {'States':['Abuja', 'Abia', 'Adamawa', 'Bauchi', 'Calabar', 'Delta', 'Jigawa', 'Kano', 'Kaduna', 'Katsina'],
              'Zones':['NC', 'SE', 'NE', 'NE', 'SS', 'SS', 'NW', 'NW', 'NW', 'NW'],
             'NumOfLGA':[17, 23, 19, 27, 12, 22, 44, 30, 29, 34],
             'NumOfWards':[136, 167, 148, 209, 113, 184, 479, 391, 196, 423],
             'Population':[20000, 27000, 22000, 31000, 11500, 23800, 47120, 41303, 36088, 44750],}

popData = pd.DataFrame(population)
popData
```

```python
vLine = popData['NumOfWards']
vLine.plot()
```

vertical bar chart

```python
popData.plot(x='States', kind='bar', stacked=True, title='Population Census')
```

horizontal bar chart

```python
zoneCount = popData.value_counts(popData['Zones'].values, sort='True')
zoneCount.plot.barh()
```
