```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
honi = pd.read_csv('honey.csv')
honi.head()
```

```python
sns.relplot(data=honi,
            x='year',y='stocks',
            col='state',
            col_wrap=3,
            kind='line')
```

```python
fig, ax = plt.subplots(figsize=(12,6))
lineplot = sns.lineplot(x=honi['year'], y=honi['numcol'], data=honi, 
                        label = 'Number of bee colonies', legend=False)
sns.despine()
plt.ylabel('# colonies')
plt.title('Price vs number of bee colonies over the years');

ax2 = ax.twinx()
lineplot2 = sns.lineplot(x=honi['year'], y=honi['priceperlb'], ax=ax2, color="r", 
                         label ='Price per lb', legend=False) 
sns.despine(right=False)
plt.ylabel('price per lb')
ax.figure.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,6))
lineplot = sns.lineplot(x=honi['year'], y=honi['priceperlb'], data=honi, 
                        label = 'Price per lb', legend=False)
sns.despine()
plt.ylabel('# Prices')
plt.title('Price vs stock over the years');

ax2 = ax.twinx()
lineplot2 = sns.lineplot(x=honi['year'], y=honi['stocks'], ax=ax2, color="r", 
                         label ='stocks', legend=False) 
sns.despine(right=False)
plt.ylabel('stocks')
ax.figure.legend()
plt.show()
```
