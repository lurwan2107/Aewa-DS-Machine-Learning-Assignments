```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
bird = pd.read_csv('birds.csv')
bird.head()
```

```python
bird.plot(kind='scatter', x='MaxLength', y='Family', figsize=(8, 8))
plt.show()
```

```python
bird['MinBodyMass'].plot(kind='hist', bins=50, figsize=(8, 8))
plt.show()
```

```python
filter = bird[(bird['MinBodyMass'] > 60) & (bird['MinBodyMass'] < 200)]
filter['MinBodyMass'].plot(kind='hist', bins=40, figsize=(8, 8))
plt.show()
```

Checking for correlation between maximum body mass and minimum length

```python
x = filter['MinBodyMass']
y = filter['MinLength']

fig, ax = plt.subplots(tight_layout=True)
disple = ax.hist2d(x, y)
plt.show()
```

```python
sns.kdeplot(filter['MaxLength'])
plt.show()
```
