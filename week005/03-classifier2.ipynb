Today's assignment is about adopting one of the classification methods to retrain the model using various parameter values.

I adopted "Ensemble classifiers" (RandomForest and AdaBoost) and passed various values to it's parameter (n_estimator).

The initial value is 100, which resulted the accuracy of 79.0% for RandomForest and 58.4% for AdaBoost. Butr when you change the values especially different values i.e if you increase the value of RandomForest and decrease the value of AdaBoost. the accuracy will increased for both methods. 

```python
import pandas as pd 
quisine = pd.read_csv('C:/Users/PC/Desktop/new_cuisine.csv')
quisine.head()
```

```python
quisine_label_df = quisine['cuisine']
quisine_label_df.head()
```

```python
quisine_feature_df = quisine.drop(['Unnamed: 0', 'cuisine'], axis=1)
quisine_feature_df.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVC 
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, precision_score,confusion_matrix, classification_report, precision_recall_curve
import numpy as np
```

```python
x_train, x_test, y_train, y_test = train_test_split(quisine_feature_df, quisine_label_df, train_size=0.3)
```

```python

classifiers = {'RFST':RandomForestClassifier(n_estimators=150),
               'ADA':AdaBoostClassifier(n_estimators=50)}


n_classifier = len(classifiers)
for index, (name, classifier) in enumerate(classifiers.items()):
    classifier.fit(x_train, np.ravel(y_train))

    pred = classifier.predict(x_test)
    accuracy = accuracy_score(y_test, pred)

    print('Accuracy (train) for %s: %0.1f%%' % (name, accuracy*100))
    print(classification_report(y_test, pred))
```
